```python
# Check the presence of missing values and delete the Columns where more than 60% data are missing.
print (df.isin([' ','NULL',0]))
print (df.isin([' ','NULL',0]).mean())
df = df.loc[:, df.isin([' ','NULL',0]).mean() < .6]
len(df.columns)


# for rows
# df2 = df[df.isnull().sum(axis=1) < 48]
# df2.shape
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
# for col_name in df.columns:
#     if(df[col_name].dtype == 'object'):
#         df[col_name] = df[col_name].astype('category')
#         df[col_name] = df[col_name].cat.codes

# df

df = pd.get_dummies(df,drop_first=True)
df
```

```python
# dependent and independent variable extraction
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
```

```python
# splitting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
# Feature Scalling --- Standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# SVM - Support Vector Machine
from sklearn import svm
model = svm.SVR(kernel='linear')
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
# SVM with Default parameter means C=1.0, kernel=rbf and gamma=auto among other parameters)
from sklearn.metrics import r2_score, mean_squared_error
print("R2 score: ", r2_score(Y_test, Y_pred))
print("MSE: ", mean_squared_error(Y_test, Y_pred))
```

```python
# SVM with rbf kernel and C=100.0
model = svm.SVC(kernel='rbf', C=100)
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
print("R2 score: ", r2_score(Y_test, Y_pred))
print("MSE: ", mean_squared_error(Y_test, Y_pred))
```

```python
# SVM with rbf kernel and C=1000.0
model = svm.SVC(kernel='rbf', C=1000)
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
print("R2 score: ", r2_score(Y_test, Y_pred))
print("MSE: ", mean_squared_error(Y_test, Y_pred))
```

```python
# SVM with linear kernel and C=1.0
model = svm.SVC(kernel='linear', C=1)
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
print("R2 score: ", r2_score(Y_test, Y_pred))
print("MSE: ", mean_squared_error(Y_test, Y_pred))
```
